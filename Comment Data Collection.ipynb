```python
#导入第三方库
import re
import json
import xlwt
import pprint
import requests
import pandas as pd
```

```python
def read_json(name):
    global content_jd,star_jd
    data_jd = json.load(open(name,'r',encoding="utf-8-sig"))['comments']
    for i in range(len(data_jd)):
        content_jd.append(re.sub(r'''\s|[\u200d]
            |[(?•ˇ?ˇ•?)!@#$%^&*
            ()！@#￥%……&*（）——-]''','',data_jd[i]['content']))
        star_jd.append(data_jd[i]['score'])
    return content_jd,star_jd
```

```python
#将预处理过的数据保存为xlsx文件
def Bookexcel(name,content,star):
    b = xlwt.Workbook(encoding = 'utf-8') #新建一个excel文件
    ws = b.add_sheet('first') #添加一个新表，名字为first
    ws.write(0,0,'content')
    ws.write(0,1,'star')
    row = 1 #写入的起始行
    #通过row和col的变化实现指向单元格位置的变化
    for i in range(len(content)):
        ws.write(row, 0 ,content[i])#向Excel文件中写入每一项
        ws.write(row, 1 ,star[i])
        row += 1
    b.save(name) 
```

```python
#京东数据采集和预处理
content_jd = []
star_jd = []
# pprint.pprint(read_json('content.json'))
for i in range(11):
    read_json('content ('+str(i)+').json')
print(content_jd,star_jd)
# 保存
Bookexcel('京东apple.xlsx',content_jd,star_jd)
df = pd.read_excel('京东apple.xlsx')
```

```python
#苏宁易购数据的采集与预处理
#好评
good_url = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-good-1-default-10-----reviewList.htm?callback=reviewList'
good_url1 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-good-3-default-10-----reviewList.htm?callback=reviewList'
good_url2 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-good-5-default-10-----reviewList.htm?callback=reviewList'
good_url3 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-good-6-default-10-----reviewList.htm?callback=reviewList'
#中评
normal_url = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-normal-1-default-10-----reviewList.htm?callback=reviewList'
normal_url1 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-normal-2-default-10-----reviewList.htm?callback=reviewList'
normal_url2 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-normal-3-default-10-----reviewList.htm?callback=reviewList'
normal_url3 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-normal-4-default-10-----reviewList.htm?callback=reviewList'
#差评
bad_url = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-bad-1-default-10-----reviewList.htm?callback=reviewList'
bad_url1 ='https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-bad-2-default-10-----reviewList.htm?callback=reviewList'
bad_url2 = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38054528-000000012314364377-0000000000-bad-3-default-10-----reviewList.htm?callback=reviewList'
content = []
star = []
def sndata(url):
    global content,star
    headers = {
    #浏览器基本信息
    'User-Agent':'''Mozilla/5.0 (Windows NT 10.0; Win64;
                     x64) AppleWebKit/537.36 (KHTML, lik
                     e Gecko) Chrome/102.0.0.0 Safari/537.36'''
    }
    r = requests.get(url,headers)
#     print(r.text)
    json_str = re.search("\(+(.*?)\)+", r.text).group(1)
    data0  = json.loads(json_str)
    data = data0['commodityReviews']
    for k in range(len(data)):
        print(data[k]['content'],data[k]['qualityStar'])
        content.append(data[k]['content'])
        star += str(data[k]['qualityStar'])
    return content,star
sndata(good_url)
sndata(good_url1)
sndata(good_url2)
sndata(good_url3)

sndata(normal_url)
sndata(normal_url1)
sndata(normal_url2)
sndata(normal_url3)

sndata(bad_url)
sndata(bad_url1)
sndata(bad_url2)
Bookexcel('苏宁易购apple.xlsx',content,star)
```
